
LHC
===

The LHC diagram.



```python
import matplotlib.pyplot as plt

plt.rcParams['mathtext.fontset'] = 'stix'
plt.rcParams['font.family'] = 'STIXGeneral'

from feynman import Diagram

fig = plt.figure(figsize=(10.,10.))
ax = fig.add_axes([0,0,1,1], frameon=False)

diagram = Diagram(ax)
#diagram.text(.5,0.9,"LHeC", fontsize=40)

in1 = diagram.vertex(xy=(.1,.8), marker='')
in2= diagram.vertex(xy=(.1,.2), marker='')
v1 = diagram.vertex(xy=(0.4,0.7))
v2 = diagram.vertex(xy=(0.4,0.3))

v3 = diagram.vertex(xy=(.65,.5))
v4 = diagram.vertex(xy=(.65,.5))

out1 = diagram.vertex(xy=(.9,.8), marker='')
out2 = diagram.vertex(xy=(.9,.2), marker='')

Wplusout = diagram.vertex(xy=(0.8, 0.7), marker='')
Wminusout = diagram.vertex(xy=(0.8, 0.3), marker='')

lw = 5
e1 = diagram.line(in1, v1, style='line', color='blue', lw=lw, arrow_param=dict(color='blue', length=0.08, width=0.02))
p1 = diagram.line(in2, v2, style='line', color='blue', lw=lw, arrow_param=dict(color='blue', length=0.08, width=0.02))

g1 = diagram.line(v3, v1, style='wiggly')
g2 = diagram.line(v2, v3, style='wiggly')

W1 = diagram.line(v3, Wplusout,  arrow=True, style='dashed', color='green', lw=3, dashes=(4, 2),arrow_param=dict(color='green', length=0.08, width=0.02))
W2 = diagram.line(v4, Wminusout, arrow=True, style='dashed', color='green', lw=3, dashes=(4, 2),arrow_param=dict(color='green', length=0.08, width=0.02))

t3 = diagram.line(v1, out1,  color='blue', lw=lw, arrow_param=dict(color='blue', length=0.08, width=0.02))
t4 = diagram.line(v2, out2 , color='blue', lw=lw, arrow_param=dict(color='blue', length=0.08, width=0.02))

e1.text(r"$p$",fontsize=40)
p1.text(r"$p$",fontsize=40)
g1.text(r"$\gamma$",fontsize=40)
g2.text(r"$\gamma$",fontsize=40)
t3.text(r"$p^{(*)}$",fontsize=40)
t4.text(r"$p^{(*)}$",fontsize=40)


#W1.text("$W^+$", fontsize=35)
#W2.text("$W^-$", fontsize=35)

diagram.plot()
plt.savefig('plot_LHC.pdf')
plt.savefig('plot_LHC.jpg')
plt.show()
```
